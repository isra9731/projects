#**Optical Character Recognition (OCR)**

Optical Character Recognition, is a process of recognizing text inside images and converting it into an electronic form. These images could be of handwritten text, printed text like documents, receipts, name cards, etc., or even a natural scene photograph.

Object detector which extracts printed or handwritten text from an image

Keras ocr provides out of the box OCR models, with the help of pipeline builds new OCR models

```python
pip install keras_ocr
```

```python
import keras_ocr as kr
import matplotlib.pyplot as plt
```

keras-ocr will automatically download pretrained weights for the detector and recognizer.

```python
pipeline=kr.pipeline.Pipeline()
```

Gets a set of five images

```python
image_ocr=[ kr.tools.read(i) 
for i in ['/content/image_ocr1.jpg','/content/image_ocr2.jpg','/content/image_ocr3.png',
          '/content/image_ocr4.jpg','/content/image_ocr5.jpg']
]
```

Provides the length of images we are using for prediction

Each list of predictions in prediction variable is a list of (word, box) tuples.

```python
prediction_list=pipeline.recognize(image_ocr)
```

Plots the predictions

```python
fig, axs = plt.subplots(nrows=len(image_ocr), figsize=(15, 15))
for ax, image, predictions in zip(axs, image_ocr, prediction_list):
    kr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
